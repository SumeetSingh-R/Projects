```python
from selenium import webdriver
import pandas as pd
import time 
from tqdm import tqdm
```

```python
FILE_PATH_FOLDER = r"C:\Users\sumit\Dt\chromedriver.exe"
search_query = 'https://www.naukri.com/data-analyst-jobs-in-bangalore?k=Data%20Analyst&l=Bangalore'
driver = webdriver.Chrome(executable_path= r"C:\Users\sumit\Dt\chromedriver.exe")
```

```python
driver.get(search_query)
```

```python
#“Data Analyst” Job position in “Bangalore” location

df=pd.DataFrame(columns=['job-title', 'job-location', 'company_name', 'experience_required','url'] )#Creating empty df

#Getting the required data
for lst in tqdm(driver.find_elements_by_css_selector(".jobTuple.bgWhite.br4.mb-8")):  
    
    try:
        jt=lst.find_element_by_css_selector("a.title.fw500.ellipsis").text
    except:
        jt ="Nan"
    try:
        jl=lst.find_element_by_css_selector(".fleft.grey-text.br2.placeHolderLi.location").text
    except:
        jl="NaN"
    try:
        cn=lst.find_element_by_css_selector("a.subTitle.ellipsis.fleft").text
    except:
        cn="Nan"
    try:
        er=lst.find_element_by_css_selector(".fleft.grey-text.br2.placeHolderLi.experience").text
    except:
        er="NaN"
    try:
        url=lst.find_element_by_tag_name('a').get_attribute("href")
    except:
        url='NaN'

#Putting those data into dataframe
    df.loc[len(df)]=[jt,jl,cn,er,url]
```
