```python
#Making a copy of data set
df1=df.iloc[:,1:]
df1.shape
#Removing outliers
from scipy.stats import zscore
z=np.abs(zscore(df1))
threshold=3
np.where(z>3)
df1=df1[(z<3).all(axis=1)]
df1.shape
```

```python
#model building

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.metrics import r2_score
```

```python
target=df1.iloc[:,7:8]
print(target)
data=df1.iloc[:,0:7]
print(data)
```

```python
x=data
x.shape
```

```python
y=target
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=50)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#using linear regression model because it contains continuous variables
lrm=LinearRegression()
lrm.fit(x_train,y_train)
print(lrm.coef_)
print(lrm.intercept_)
score=lrm.score(x_train,y_train)
print(score)
pred=lrm.predict(x_test)
print('predicted result:',pred)
print('actual result:',y_test)
```

```python
#using Random forest Regressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
regr = RandomForestRegressor()
regr.fit(x_train,y_train)
score=regr.score(x_train,y_train)
print(score)

pred=regr.predict(x_test)
print('predicted result:',pred)
print('actual result:',y_test)
```

```python
#Using SVR
from sklearn.svm import SVR
svr = SVR()
svr.fit(x_train,y_train)
score=svr.score(x_train,y_train)
print(score)
pred=svr.predict(x_test)
print('predicted result:',pred)
print('actual result:',y_test)
```

```python
#using Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(x_train,y_train)
score=dtr.score(x_train,y_train)
print(score)
pred=dtr.predict(x_test)
print('predicted result:',pred)
print('actual result:',y_test)
```

```python
#Saving the model
import joblib

#Save the model as a pickle in a file
joblib.dump(dtr,'abalone.pkl')
```
